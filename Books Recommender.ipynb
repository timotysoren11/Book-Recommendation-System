```python
#Removing duplicates
final_rating.drop_duplicates(['user_id', 'title'], inplace = True)
```

# Performing Pivot-table on dataset for finding out the similar interest of users - Clustering dataset

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot 
```

```python
book_pivot.shape
```

```python
#Removing NaN value and replace with Zeros

book_pivot.fillna(0, inplace=True)
```

```python
book_pivot 
```

```python
#For ignoring the value zeros
# Performing csr Matrix

from scipy.sparse import csr_matrix 
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
# importing cluster algorithm
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
# finding the index and suggestion of the books
# 237 is the index of the book

distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
# see the index ...books recommended
suggestion
```

```python
# checking the name of the book
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

# Testing the recommendation system

```python
# Running the function
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
     
```

```python
book_name = 'A Bend in the Road'
recommend_book(book_name)
```

```python
book_name = 'Without Remorse'
recommend_book(book_name)
```
